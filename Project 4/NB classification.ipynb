```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df.gender = [1 if i == "Male" else 0 for i in df.gender]
```

```python
df.gender
```

```python
df.head()
```

```python
x = df.drop(["gender"], axis=1)

y= df.gender.values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train, y_train)

print("print training accuracy of NBC algo :",gnb.score(x_train,y_train))
print("print testing accuracy of NBC algo :",gnb.score(x_train,y_train))
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb =  MultinomialNB()

mnb.fit(x_train, y_train)

print("print training accuracy:",mnb.score(x_train,y_train))
print("print testing accuracy:",mnb.score(x_train,y_train))
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb= BernoulliNB()

bnb.fit(x_train, y_train)

print("print training accuracy:",bnb.score(x_train,y_train))
print("print testing accuracy:",bnb.score(x_train,y_train))
```
